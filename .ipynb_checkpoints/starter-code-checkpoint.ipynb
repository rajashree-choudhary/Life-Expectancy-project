2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
#Code for Standard Deviation

def mean(values):
    return sum(values)/len(values)

def standard_deviation(values):
    mean_e = mean(values)
    variance = sum((i - mean_e)**2 for i in values)/len(values)
    return variance ** 0.5

numbers = [1, 2, 3, 4]
print(standard_deviation(numbers))
```

```python
#changing M and K to numeric value

def convert_suffix_to_numeric(value):
    if isinstance(value, float) or isinstance(value, int):
        return float(value)
    elif 'M' in value:
        return float(value.replace('M', '')) * 1e6
    elif 'k' in value:
        return float(value.replace('k', '')) * 1e3
    else:
        return float(value)

# Apply the conversion function to the entire DataFrame
population.iloc[:, 1:] = population.iloc[:, 1:].applymap(convert_suffix_to_numeric)

# Display the updated DataFrame
print(population)
```

```python
# Code

def histogram_draw(feature_name, plot_title):
    plt.hist(gapminder_1989[feature_name], bins=20, alpha=0.5, label='1989')
    plt.hist(gapminder_2023[feature_name], bins=20, alpha=0.5, label='2023')
    plt.title(plot_title)
    plt.xlabel(feature_name)
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.show()
histogram_draw('life_exp', 'Life expectancy distribution in 1989 and 2023')
```

```python
#Write a function to draw a box plot for each feature
def box_draw(feature_name, plot_title_1, plot_title_2, log_scale=False):
    gapminder_1989.boxplot(feature_name, by='continents', rot=90)
    plt.title(plot_title_1)
    plt.suptitle("")
    plt.xlabel('Continent')
    plt.ylabel(feature_name)
    if log_scale: plt.yscale('log')
        
    gapminder_2023.boxplot(feature_name, by='continents', rot=90)
    plt.title(plot_title_2)
    plt.suptitle("")
    plt.xlabel('Continent')
    plt.ylabel(feature_name)
    if log_scale: plt.yscale('log')
```

```python
box_draw('life_exp', 'Life expectancy in 1989', 'Life expectancy in 2023');
```

```python
# Create a list of 5 different countries
five_countries = ['Colombia', 'Nigeria', 'Germany', 'India', 'Australia']

# Write a function to return a dataframe for a country
def make_dataframe_for_country(country_name):
    df = pd.DataFrame()
    df['life_expectancy'] = life_exp.query('country == @country_name').loc[:, '1800':'2023'].transpose().iloc[:, 0]  
    df['income_per_person'] = income_pp.query('country == @country_name').loc[:, '1800':'2023'].transpose().iloc[:, 0]
    df['population'] = population.query('country == @country_name').loc[:, '1800':'2023'].transpose().iloc[:, 0]
    df['child_mortality'] = child_mortality.query('country == @country_name').loc[:, '1800':'2023'].transpose().iloc[:, 0]
    df['children_per_woman'] = children_per_women.query('country == @country_name').loc[:, '1800':'2023'].transpose().iloc[:, 0]
    df.index = pd.to_datetime(df.index)  # convert index to datetime
    return df

five_countries_df = list(map(make_dataframe_for_country, five_countries))
```

```python
# Writing a function to draw a plot for a feature
def draw_one_feature(feature_name, country_df, country_name):
    plt.plot(country_df[feature_name], label=country_name)
    plt.title(feature_name + ' from 1800 to 2023 for all countries')
    plt.legend(loc='upper left')

# Get the list of feature names
feature_list = five_countries_df[-1].columns

# Use map to apply the draw_one_feature function to each feature
for feature_name in feature_list:
    plt.figure(figsize=(6, 4))
    list(map(lambda country_df, country_name: draw_one_feature(feature_name, country_df, country_name), five_countries_df, five_countries))

plt.show()
plt.savefig('../plots/countries.png');
```
